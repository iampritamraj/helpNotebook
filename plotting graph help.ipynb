```python
# Set up the data
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2, 6, 14, 30, 43, 75]
# Set the axes headings
pyplot.ylabel('Speed', fontsize=12)
pyplot.xlabel('Time', fontsize=12)
# Set the title
pyplot.title("Speed v Time")
# Plot and display the graph
# Using blue circles for markers ('bo')
# and a solid line ('-') 
pyplot.plot(x, y, 'bo-')
pyplot.show()
```

```python
time = [1,2,3,4,5,6,7,8,9]
speed = [25,30,35,40,45,48,485,284,293]
```

```python
pyplot.xlabel("Time",fontsize=15,color = "blue")
pyplot.ylabel("Speed",fontsize = 15,color = "red")
pyplot.title("Speed VS Time",fontsize= 15,color = "green")
pyplot.legend("Time")
pyplot.plot(time,speed,c='red',)
```
