```python
import urllib
url = "http://spys.me/proxy.txt"
file = urllib.request.urlopen(url)
text = []
for line in file:
	decoded_line = line.decode("utf-8")
	text.append(decoded_line)
```

```python
text
```

```python
import re
import pandas as pd
ips = []
for i in text[9:-4]:
    ip = i.split()[0]
    ips.append(ip)
ips
```

```python
df = pd.DataFrame(data=ips)
 
```

```python
print("please enter filename as u want")
name = input()

```
